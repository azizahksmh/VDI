```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

Link Dataset :

https://www.kaggle.com/datasets/iqrar99/lokasi-mudik-karyawan

```python
dataa = pd.read_csv('/content/catatan_lokasi.csv')
dataa.head(5)
```

```python
dataa.isnull().sum()
```

```python
dataa['lokasi_dominan'].unique()
```

```python
dataa['lokasi_dominan'] = dataa['lokasi_dominan'].replace({'Kota ': '', 'Kabupaten ': ''}, regex=True)
```

```python
kota_counts = dataa['lokasi_dominan'].value_counts()
kota_percentage = dataa['lokasi_dominan'].value_counts(normalize=True) * 100
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=kota_percentage.index, y=kota_percentage.values, palette='viridis')
plt.title('Persentase Kota Asal Karyawan')
plt.xlabel('Kota Asal')
plt.xticks(rotation=90)
plt.ylabel('Persentase (%)')
plt.show()
```

```python
plt.figure(figsize=(14, 12))
sns.barplot(x=kota_percentage.index, y=kota_percentage.values, palette='viridis')
for index, value in enumerate(kota_percentage.values):
    plt.text(index, value + 0.5, f'{value:.1f}%', ha='center')
plt.title('Persentase Kota Asal Karyawan dengan Label Persentase')
plt.xlabel('Kota Asal')
plt.ylabel('Persentase (%)')
plt.xticks(rotation=90)
plt.show()
```
