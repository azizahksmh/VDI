# **Visualisasi Bar Chart**

```python
# Menghilangkan kata "Kota" dan "Kabupaten" dari dataset untuk mempersingkat penamaan pada visualisasi
dataa['lokasi_dominan'] = dataa['lokasi_dominan'].replace({'Kota ': '', 'Kabupaten ': ''}, regex=True)
```

```python
# Menghitung berapa kali suatu kota muncul dari dataset
kota_counts = dataa['lokasi_dominan'].value_counts()
# Menghitung persentase dari setiap kemunculan kota
kota_percentage = dataa['lokasi_dominan'].value_counts(normalize=True) * 100
```

```python
# Membuat visualisasi Bar Chart
plt.figure(figsize=(14, 12))
sns.barplot(x=kota_percentage.index, y=kota_percentage.values, palette='viridis', hue=kota_percentage.index, legend=False)
for index, value in enumerate(kota_percentage.values):
    plt.text(index, value + 0.5, f'{value:.1f}%', ha='center')
plt.title('Bar Chart Persentase Kota Lokasi Mudik Karyawan')
plt.xlabel('Lokasi Mudik Karyawan')
plt.ylabel('Persentase (%)')
plt.xticks(rotation=90)
plt.show()
```
